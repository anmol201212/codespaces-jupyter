# Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a Normal Distribution (1-D)

u = 5
sigma = 2

vals = np.random.randn(1000)
vals = u + sigma*vals
print(vals.shape)
print(vals)
```

```python
plt.hist(vals,100)
plt.show()
```
